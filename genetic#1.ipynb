```python
import random as rand
import math as math
```

<h4 style="color:gold;"> Utility Functions : </h4>

```python
# convert int set to binary set
def intToBinary(int):
    trimed = format(int, 'b')
    

    while(len(trimed) < 5):
        trimed = "0" + trimed

    return trimed

def convertIntSetTobinary(parent_set):
    binaryParent = []
    for parent in parent_set:
        binaryParent.append(intToBinary(parent))
        #print (parent, " => binary " ,intToBinary(parent))
    return binaryParent
```

```python
#creating early Parents
def earlyMatingPool(count):
    parent_set = set()
    while (len(parent_set) < count):
        random = rand.randint(0,30)
        parent_set.add(random)
    return parent_set
```

<h4 style="color:gold;"> Fitness Function : </h4>

```python
#GET : X      RETURN : [Y,X]
def fitness(Xset):
    result = []
    for x in Xset:
        conc = (-(x ** 2)+ 6*x)
        result.append([conc,x]) # result == > [y,x]
    return result
```

<h4 style="color:gold;"> Parent Selection Method : </h4>

```python
def parentSelection(qualified):
    selected = []
    # x = []
    # y = []
    # for x_set in qualified:
    #     x.append(x_set[1])
    # for y_set in qualified:
    #     y.append(y_set[0])
    for i in range(round(len(qualified)/2)):
        current_max = max(qualified)
        selected.append(current_max)
        qualified.remove(current_max)

    return selected
```

```python
def binaryToInt(string):
    return int(string,2)
```

<h4 style="color:gold;"> Probablity executer : </h4>

```python
def accu():
    number = rand.randint(0,100)
    if(number > 50):
        return True
    else:
        return False
```

```python
def eventProb(prob):
    number = rand.randint(0,100)
    if(number <= prob):
        return True
    else:
        return False
```

<h4 style="color:gold;"> Muatation Proccess : </h4>

```python
def mutate(candidate,mutate_prob):
    next_gen = []
    for parent in candidate:
        offspring = ''
        if(eventProb(mutate_prob)):
            #mutation will happen here 
            seed = intToBinary(parent)
            for i in range(0,len(seed)):
                if(not accu()):
                    offspring += seed[i]
                else:
                    if(seed[i] == '0'):
                        offspring += '1'
                    else:
                        offspring += '0'
            next_gen.append(binaryToInt(offspring))
        else:
            next_gen.append(parent)
    return next_gen
```

<h4 style="color:gold;"> Cross Over Process (1,2) : </h4>

```python

def combination(candidates,cross_p,cross_method):
    counter = math.ceil(len(candidates)/2)
    next_generation = []
    offspring1 = ''
    offspring2= ''
    for i in range(0,counter):
        if (len(candidates) == 1):
            next_generation.append(candidates[0][1])
            break  
        rand.shuffle(candidates)
        print ('After Shuffle :',candidates,'\n')
        print('\n two selected  ===> ',candidates[-1],candidates[-2])
        seed1 = intToBinary(candidates[-1][1])
        seed2 = intToBinary(candidates[-2][1])
        del candidates[-1]
        del candidates[-1]
        if(eventProb(cross_p) == True):
            #choosing cross over 1 point
            if (cross_method == 1):
                cross_over_point = rand.randint(0,len(seed1))
                offspring1 = seed1[0:cross_over_point] + seed2[cross_over_point:len(seed2)]
                offspring2 = seed2[0:cross_over_point] + seed1[cross_over_point:len(seed1)]
            elif(cross_method == 2):
                offspring1=''
                offspring2=''
                coin_flip = False
                for i in range(0,len(seed1)):
                    coin_flip = accu()
                    if(coin_flip == True):
                        offspring1 = offspring1 + seed1[i]
                    else:
                        offspring1 = offspring1 + seed2[i]
                offspring2 = ''
                for k in range(0,len(offspring1)):
                    if(offspring1[k] == '0'):
                        offspring2 += '1'
                    elif(offspring1[k] == '1'):
                        offspring2 += '0'
            next_generation.append(binaryToInt(offspring1))
            next_generation.append(binaryToInt(offspring2))
        else:
            next_generation.append(binaryToInt(seed1))
            next_generation.append(binaryToInt(seed2)) 
    return next_generation
```

<h4 style="color:gold;"> Genetic Algoritm : </h4>

```python
files = open('config.txt')
configs = files.readlines()
# 0 => First Total population 
# 1 => Maximum amount of iteration 
# 2 => Prob of mutation
# 3 => Prob of crossover
# 4 => Method of crossover : [1 or 2]
count_early_pop_parents = int(configs[0])
mux_iteration = int(configs[1])
generation_number = 0

int_early_pop_parents = earlyMatingPool(count_early_pop_parents) #initalize
binary_early_pop_parents = convertIntSetTobinary( int_early_pop_parents) 
print('All First Parents : ',int_early_pop_parents, ' #random selection \n')
evaluated = fitness(int_early_pop_parents)
selected_parent = parentSelection(evaluated) #evaluete and select parents
print('selected Parents ===>',selected_parent, ' \n')

while (mux_iteration > generation_number):
    print('\n--------------------------------------------------------------\n')
    print('\n\n         [interation number :::: ' , generation_number+1,']\n\n')
    newCandidate = combination(selected_parent,int(configs[3]),int(configs[4]))
    print('After Cross Over  () ===>  ',newCandidate)
    mutatedCandidate = mutate(newCandidate,int(configs[2]))
    print('After Mutation    () ===>  ',mutatedCandidate)
    selected_parent = fitness(mutatedCandidate)
    print('\nfitness : [y,x] : ',selected_parent)
    print('\n--------------------------------------------------------------\n')
    generation_number+= 1
    
    

print(' \n\n\n\n*** RESULT   : \n')
print('All First Parents : ',int_early_pop_parents, ' \n')
print('Last genertaion : ',newCandidate, ' \n')
final_result = max(selected_parent)
print('\n\nThe Mux Result Is :::::::::: ',final_result)
```
